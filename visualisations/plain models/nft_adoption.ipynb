```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
import random


data = pd.read_csv('/Users/aoamacsplace/Documents/Projects/Online retail/data-models/simulated_data.csv')   
```

```python
def analyze_nft_adoption(data, min_total_nft, max_total_nft, min_rewards_nft, max_rewards_nft):
    
    data = data[(data['totalNft'] >= min_total_nft) & (data['totalNft'] <= max_total_nft)]
    data = data[(data['rewardsNft'] >= min_rewards_nft) & (data['rewardsNft'] <= max_rewards_nft)]
    data['nft_adoption_score'] = data['op_tx_nft'] * data['reputation_score']
    data = data.sort_values('nft_adoption_score', ascending=False)
    
    return data

# to use
min_total_nft = 45
max_total_nft = 50
min_rewards_nft = 10
max_rewards_nft = 10

nft_adoption_data = analyze_nft_adoption(data, min_total_nft, max_total_nft, min_rewards_nft, max_rewards_nft).reset_index(drop=True)
nft_adoption_data = pd.DataFrame(nft_adoption_data, columns=['ensName','op_tx_nft', 'reputation_score', 'nft_adoption_score'])
print(nft_adoption_data)
```

```python
nft_adoption_data = pd.DataFrame(nft_adoption_data, columns=['ensName', 'op_tx_nft', 'reputation_score', 'nft_adoption_score'])

fig = px.bar(nft_adoption_data, x='ensName', y='nft_adoption_score', 
             hover_data=['op_tx_nft', 'reputation_score'],
             labels={'ensName': 'ENS Name', 'nft_adoption_score': 'NFT Adoption Score'},
             title='NFT Adoption by ENS Names')

fig.update_layout(xaxis_tickangle=-45)

fig.show()
```
